```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
df_conf = pd.read_csv('corona_data/time_series_covid_19_confirmed.csv')
df_death = pd.read_csv('corona_data/time_series_covid_19_deaths.csv')
df_recovered = pd.read_csv('corona_data/time_series_covid_19_recovered.csv')
```

```python
df_conf = df_conf.groupby('Country/Region').sum().drop(['Lat', 'Long'], axis=1).reset_index()
df_conf
```

```python
df_death = df_death.groupby('Country/Region').sum().drop(['Lat', 'Long'], axis=1).reset_index()
df_death
```

```python
df_recovered = df_recovered.groupby('Country/Region').sum().drop(['Lat', 'Long'], axis=1).reset_index()
df_recovered
```
